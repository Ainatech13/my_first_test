##Exercice
stantardiser la matrice A sur chaque colonne.
A - mean(A) / ecrat-type(A) 
resultat : chaque colonne doit avoir une moyenne = 0 et une ecart type = 1
standardisation des colonnes  independament l'une de l'autre 

```python
import numpy as np
np.random.seed(0) 
A = np.random.randint(0, 100, [10, 5] ) 

print(A)
```

## solution 
A - mean(A) / ecrat-type(A) 
# mean selon la colonne = 
for i in range(A.shape[1]+1 ) 
B = 
mean = np.mean(axis= 0) 

```python
Z = np.zeros_like(A, dtype=float) 
for i in range(0, A.shape[1]):
    col_A= A[:, i:i+1]
    mean_A = col_A.mean(axis= 0)
    std_A = col_A.std(axis = 0) 
    Z[:, i]   = ((col_A - mean_A )/ std_A).ravel()

print(Z)     
    
    
 
```

## exercise matplotlib 

```python
dataset = {f"experience {i}" : np.random.randn(100) for i  in range(4) } 

dataset
```

```python
k = len(dataset) 
def graphique(dataset):
    plt.figure() 
    for i in range(k) :
        plt.subplot(k, 1, i+1) 
        plt.plot(dataset[f"experience {i}"])
    plt.show()

graphique(dataset)
             
    
    
```

## version subplots : 

```python
k = len(dataset) 
def graphique(dataset):
    
    fig, ax = plt.subplots(k, 1)     
    for i in range(k) :
        ax[i].plot(dataset[f"experience {i}"]) 
    plt.show()
        
    

graphique(dataset)
```
