```python
def e_potenitelle(masse, hauteur, e_limite= 3000, g=9.81): 
    E = masse*hauteur*g
    print(E, 'joules') 
    if E < e_limite :
        print('energie basse ') 
    elif E >= e_limite : 
        print('energie haute') 
        
    
    return E 

e_potenitelle(500, 1000) 


```

```python
#suite de fibonacci : 
#suite de fibonacci jusqu'a une valeur de n 
def fibonaci(n) :
    a = 0
    liste=[a]
    b= 1
    
    
    while a < n :
        print(a)
        a , b = b, a+b
        liste.append(a)
    return liste


print(fibonaci(40))
        
        

    






        
    
    
```
