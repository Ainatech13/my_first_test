```python
#normalisation simple : 
liste= [1, 2, 3, 4, 5]
def normalisation(liste) :
    min = liste[0]
    max = liste[0]
    #recherche de min et max :
    for x in liste :
        if x < min : 
            min = x 
        elif x > max :
            max = x

    #normalisation de x :
    x_norm = (x-min)/(max-min)

    return x_norm 
    
            
     
    


print(normalisation(liste))
    

```

```python
##min, max , moyenne, variance : 
data = [17, 16, 20, 18, 18, 19, 19, 20, 21, 22, 17, 18]

def analyse(data) :
    #recherche min et max :
    min = data[0]
    max = data[0]
    for x in data :
        if x>max :
            max = x
        elif x < min : 
            min = x

    #calcul moyenne :
    sum = 0
    for i in data:
        sum = sum+i
    moyenne = sum / len(data)
    #calcul variance :
    for x_i in data : 
        variance = ((x_i - moyenne)**2 )/ len(data)
    return min, max, moyenne, variance

print(analyse(data))
```

```python
labels = ["A", "B", "A", "C", "D"]

def unique(labels): 
    uniques = []
    for label in labels :
        if label not in uniques : 
            uniques.append(label)
    return uniques

print(unique(labels))


#clean data to binary data :
def binary_data(labels): 
    categories = unique(labels)
    result=[]
    for label in labels :
        vecteur = []
        for cat in categories : 
            if label == cat : 
                vecteur.append(1)
            else :
                vecteur.append(0) 
        result.append(vecteur)

    return result 

print(binary_data(labels))

    
            
```
